**그래프 해석**
도미는 길이가 변하면 무게가 많이 무거워지지만 빙어는 길이가 조금 길어졌다고 무게가 확연히 늘지 않는 다는것을 볼 수 있음 즉 빙어는 길이와 무게의 상관관계가 낮음

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data=[[l,w] for l,w in zip(length,weight)]
```

```python
fish_data
```

```python
fish_target = [1] * 35 + [0] * 14
fish_target
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier()
```

```python
kn.fit(fish_data,fish_target)
```

```python
kn.score(fish_data,fish_target)
```

```python
kn.predict([[30,600]])
```

```python
kn._fit_X
```

```python
kn._y
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
print(35/49)
```

**고찰**
knn알고리즘은 최근접 이웃 알고리즘이다. 디폴트 값은 5인데 이 뜻은 가까운 5개의 데이터를 가지고 값을 예측하는 것이다. 따라서 kn49는 이웃한 데이터 49개를 가지고 데이터를 예측하는 것인데 kn에 사용한 데이터개수가 49개이고 도미의 데이터가 35개니까 전부다 도미라고 예측할것이고 그럼 49개중에 35개는 맞출거니까 25줄과25줄의 값이 같은것이다.
